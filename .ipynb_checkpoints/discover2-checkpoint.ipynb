```python
%run ppo.py
%run discriminator.py

np.set_printoptions(suppress = True)
logger = Logger()
ppo = PPO(logger, parallel = 1)
discrim = SmallD(logger, s = 11, a = 3)

def rollout_real_ppo(agent, discrim, env, batch_size = 50000):
    
    state = env.reset()
    rewards, fake_rewards = 0,0
    eps_rewards, eps_fake_reward = [], []
    for i in range(batch_size):
        action = agent.select_action(state)

        next_state, reward, done, _ = env.step(action.flatten())
        fake_reward = discrim(state.reshape(1,-1),action.reshape(1,-1)).detach().item() 
        
        agent.buffer.rewards.append(reward)
        agent.buffer.is_terminals.append(done)
        rewards += reward
        fake_rewards += fake_reward
        if done:
            eps_rewards.append(rewards)
            eps_fake_reward.append(fake_rewards)
            rewards, fake_rewards = 0, 0
            state = env.reset()
            continue 
            
        state = next_state
    print('mean real reward', np.asarray(eps_rewards).mean())
    print('mean fake reward', np.asarray(fake_rewards).mean())
    
def train_discrim(e_states, e_actions, l_states, l_actions, num_train = 20, bs = 256):
    losses = []
    
    for _ in range(num_train):
        idx = np.random.permutation(l_states.shape[0])[:bs]
        e_idx = np.random.permutation(e_states.shape[0])[:bs]
        #e_idx = geometric_index(e_states.shape[0], bs)

        l_state_batch, l_action_batch = l_states[idx], l_actions[idx]
        e_state_batch, e_action_batch = e_states[e_idx], e_actions[e_idx]
        loss = discrim.step(l_state_batch, l_action_batch, e_state_batch, e_action_batch)
    
        losses.append(loss)
    return losses  

for i in range(10):
    rollout_real_ppo(ppo, discrim, env)
    train_discrim(e_states, e_actions, torch.stack(ppo.buffer.states, dim=0), torch.stack(ppo.buffer.actions, dim=0))
    ppo.update()
    

    
```
